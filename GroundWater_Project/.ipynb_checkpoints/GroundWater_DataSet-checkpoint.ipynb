```python
import pandas as pd
import numpy as np
import random

data = []

seasons = ["summer", "monsoon", "winter"]

for _ in range(1000):
    tank_capacity = random.randint(5000, 20000)
    current_level = random.randint(1000, tank_capacity)
    residents = random.randint(5, 50)
    
    base_usage = residents * random.uniform(120, 180)
    
    rainfall = random.uniform(0, 10)
    season = random.choice(seasons)
    
    if season == "summer":
        usage = base_usage * 1.2
    elif season == "monsoon":
        usage = base_usage * 0.9
    else:
        usage = base_usage
    
    usage = usage * (1 - rainfall * 0.02)
    
    days_left = current_level / usage
    
    data.append([
        tank_capacity,
        current_level,
        residents,
        usage,
        rainfall,
        season,
        round(days_left, 2)
    ])

df = pd.DataFrame(data, columns=[
    "tank_capacity",
    "current_level",
    "residents",
    "daily_usage",
    "rainfall",
    "season",
    "days_left"
])

df.to_csv("water_data.csv", index=False)

df.head(10)
```
